```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def preprocess_image(image_path, target_size=(256, 256)):
    # Read the image
    image = cv2.imread(image_path)

    # Resize the image to the target size
    image = cv2.resize(image, target_size)

    # Perform any additional preprocessing steps (e.g., cropping, normalization)
    # Example:
    #image = image[50:450, 100:500]  # Crop a specific region
    image = cv2.normalize(image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)  # Normalize

    return image

def preprocess_dataset(dataset_dir, target_size=(256, 256)):
    preprocessed_images = []
    image_paths = os.listdir(dataset_dir)

    for image_path in image_paths:
        if image_path.endswith('.jpg') or image_path.endswith('.png'):
            full_path = os.path.join(dataset_dir, image_path)
            preprocessed_image = preprocess_image(full_path, target_size)
            preprocessed_images.append(preprocessed_image)

    return preprocessed_images

def show_images(images):
    num_images = len(images)
    fig, axes = plt.subplots(1, num_images, figsize=(10, 5))

    for i in range(num_images):
        axes[i].imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        axes[i].axis('off')

    plt.show()

# Example usage:
dataset_dir = '/content/drive/MyDrive/image'
preprocessed_images = preprocess_dataset(dataset_dir)

# Show the preprocessed images
show_images(preprocessed_images)
```

```python
def preprocess_image(image_path, target_size=(256, 256)):
    image = cv2.imread(image_path)
    image = cv2.resize(image, target_size)
    return image

def preprocess_dataset(dataset_dirs, label_file, target_size=(256, 256)):
    preprocessed_images = []
    labels = []

    # Read labels from the CSV file
    label_df = pd.read_csv(label_file)
    label_dict = dict(zip(label_df['filename'], label_df['label']))

    for dataset_dir in dataset_dirs:
        image_paths = os.listdir(dataset_dir)

        for image_name in image_paths:
            if image_name.endswith('.jpg') or image_name.endswith('.png'):
                full_path = os.path.join(dataset_dir, image_name)
                preprocessed_image = preprocess_image(full_path, target_size)
                preprocessed_images.append(preprocessed_image)
                labels.append(label_dict[image_name])

    return preprocessed_images, labels

# Function to visualize images with labels
def visualize_images(images, labels, num_images=5):
    num_rows = (num_images // 5) + 1
    fig, axes = plt.subplots(num_rows, 5, figsize=(15, 3*num_rows))

    for i, ax in enumerate(axes.flat):
        if i < len(images):
            ax.imshow(images[i])
            ax.axis('off')
            ax.set_title(labels[i])
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage:
dataset_dirs = ['/content/drive/MyDrive/image']  # List of dataset directories
label_file = '/content/drive/MyDrive/annotations.csv'
preprocessed_images, labels = preprocess_dataset(dataset_dirs, label_file)

# Visualize the preprocessed images with their labels
visualize_images(preprocessed_images, labels, num_images=10)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(preprocessed_images, labels, test_size=0.2, random_state=42)

# Now, X_train and y_train contain the training images and their labels respectively,
# while X_test and y_test contain the testing images and their labels respectively.
```

```python
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential([
        # Convolutional layers
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        # Flatten layer
        layers.Flatten(),

        # Dense (fully connected) layers
        layers.Dense(128, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

# Example usage:
input_shape = (256, 256, 3)  # Assuming input images are resized to 256x256 and have 3 channels (RGB)
num_classes = 3  # Assuming you have 3 classes: 'new', 'used', 'torn_cover'

# Create the CNN model
cnn_model = create_cnn_model(input_shape, num_classes)

# Print model summary
cnn_model.summary()
```
